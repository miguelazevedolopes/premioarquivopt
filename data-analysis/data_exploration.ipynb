## Frequency of mentions of other political groups in each one

```python
from queue import Empty


def political_groups_frequency(df, groups):

    df["text"] = df['text'].str.replace(r'[^\w\s]','',regex=True)    
    
    word_frequency = df.text.str.split().explode().value_counts().reset_index()
    
    word_frequency.columns = ['Word', 'Frequency'] 
    
    frequency0 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[0] + "$", case=False, flags=re.IGNORECASE)]
    frequency1 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[1] + "$", case=True)]
    frequency2 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[2] + "$", case=True)]  
    frequency3 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[3] + "$", case=True)]
    frequency4 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[4] + "$", case=False, flags=re.IGNORECASE)]
    frequency5 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[5] + "$", case=False, flags=re.IGNORECASE)]  
    frequency6 = word_frequency.loc[word_frequency['Word'].str.contains("^" + groups[6] + "$", case=False, flags=re.IGNORECASE)]

    if(frequency0.empty):  
        final_frequence0 = 0
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]

    elif(frequency1.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = 0
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]
    elif(frequency2.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = 0
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]
    elif(frequency3.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = 0
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]
    elif(frequency4.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = 0        
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]
    elif(frequency5.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = 0
        final_frequence6 = frequency6.iloc[0]["Frequency"]
    elif(frequency6.empty):
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = 0
    else:
        final_frequence0 = frequency0.iloc[0]["Frequency"]
        final_frequence1 = frequency1.iloc[0]["Frequency"]
        final_frequence2 = frequency2.iloc[0]["Frequency"]
        final_frequence3 = frequency3.iloc[0]["Frequency"]
        final_frequence4 = frequency4.iloc[0]["Frequency"]
        final_frequence5 = frequency5.iloc[0]["Frequency"]
        final_frequence6 = frequency6.iloc[0]["Frequency"]

    return [final_frequence0, final_frequence1, final_frequence2, final_frequence3, final_frequence4, final_frequence5, final_frequence6]
```
